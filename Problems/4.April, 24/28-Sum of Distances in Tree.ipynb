There is an undirected connected tree with `n` nodes labeled from `0` to `n - 1` and `n - 1` edges.

You are given the integer `n` and the array `edges` where **edges[i] = [a<sub>i</sub>, b<sub>i</sub>]** indicates that there is an edge between nodes **a<sub>i</sub>** and **b<sub>i</sub>** in the tree.

Return an array `answer` of length `n` where `answer[i]` is the sum of the distances between the **i<sup>th</sup>** node in the tree and all other nodes.

<br>

**Example 1:**

![Ic-sumdist1](../../images/lc-sumdist1.jpg)

>**Input:** n = 6, edges = [[0,1],[0,2],[2,3],[2,4],[2,5]]<br>
>**Output:** [8,12,6,10,10,10]<br>
>**Explanation:** The tree is shown above.<br>
>We can see that dist(0,1) + dist(0,2) + dist(0,3) + dist(0,4) + dist(0,5)<br>
>equals 1 + 1 + 2 + 2 + 2 = 8.<br>
>Hence, answer[0] = 8, and so on.

**Example 2:**

![Ic-sumdist2](../../images/lc-sumdist2.jpg)

>**Input:** n = 1, edges = []<br>
>**Output:** [0]

**Example 3:**

![Ic-sumdist3](../../images/lc-sumdist3.jpg)

>**Input:** n = 2, edges = [[1,0]]<br>
>**Output:** [1,1]

<br>

**Constraints:**
- >1 <= n <= 3 * 10<sup>4</sup>
- >edges.length == n - 1
- >edges[i].length == 2
- >0 <= a<sub>i</sub>, b<sub>i</sub> < n
- >a<sub>i</sub> != b<sub>i</sub>
- >The given input represents a valid tree.

```python
class Solution:
    def sumOfDistancesInTree(self, n: int, edges: list[list[int]]) -> list[int]:
        from collections import defaultdict


        count = [1] * n
        result = [0] * n
        
        graph = defaultdict(list)
        for u, v in edges:
            graph[u].append(v)
            graph[v].append(u)
        
        def dfs(node, parent):
            for child in graph[node]:
                if child != parent:
                    dfs(child, node)
                    count[node] += count[child]
                    result[node] += result[child] + count[child]
            
        dfs(0, -1)
        
        def dfs2(node, parent):
            for child in graph[node]:
                if child != parent:
                    result[child] = result[node] - count[child] + (n - count[child])
                    dfs2(child, node)
        
        dfs2(0, -1)
        
        return result
```
