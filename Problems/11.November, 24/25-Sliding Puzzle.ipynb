On an 2 x 3 board, there are five tiles labeled from 1 to 5, and an empty square represented by 0. A move consists of choosing 0 and a 4-directionally adjacent number and swapping it.

The state of the board is solved if and only if the board is [[1,2,3],[4,5,0]].

Given the puzzle board board, return the least number of moves required so that the state of the board is solved. If it is impossible for the state of the board to be solved, return -1.

<br>

**Example 1:**<br><br>
![image1](../../images/slide1-grid.jpg)<br>
>**Input:** board = [[1,2,3],[4,0,5]]<br>
>**Output:** 1<br>
>**Explanation:** Swap the 0 and the 5 in one move.

**Example 2:**<br><br>
![image2](../../images/slide2-grid.jpg)<br>
>**Input:** board = [[1,2,3],[5,4,0]]<br>
>**Output:** -1<br>
>**Explanation:** No number of moves will make the board solved.

**Example 3:**<br><br>
![image3](../../images/slide3-grid.jpg)<br>
>**Input:** board = [[4,1,2],[5,0,3]]<br>
>**Output:** 5<br>
>**Explanation:** 5 is the smallest number of moves that solves the board.<br>
>An example path:<br>
>After move 0: [[4,1,2],[5,0,3]]<br>
>After move 1: [[4,1,2],[0,5,3]]<br>
>After move 2: [[0,1,2],[4,5,3]]<br>
>After move 3: [[1,0,2],[4,5,3]]<br>
>After move 4: [[1,2,0],[4,5,3]]<br>
>After move 5: [[1,2,3],[4,5,0]]

<br>

**Constraints:**
- >board.length == 2
- >board[i].length == 3
- >0 <= board[i][j] <= 5
- >Each value board[i][j] is **unique**.

```python
class Solution:
    def slidingPuzzle(self, board: list[list[int]]) -> int:
        from collections import deque

        start = ''.join(str(num) for row in board for num in row)
        target = "123450"
        
        neighbors = {
            0: [1, 3],
            1: [0, 2, 4],
            2: [1, 5],
            3: [0, 4],
            4: [1, 3, 5],
            5: [2, 4]
        }
        
        queue = deque([(start, start.index('0'), 0)])
        visited = set()
        visited.add(start)
        
        while queue:
            state, zero_pos, steps = queue.popleft()
            
            if state == target:
                return steps
            
            for neighbor in neighbors[zero_pos]:
                new_state = list(state)
                new_state[zero_pos], new_state[neighbor] = new_state[neighbor], new_state[zero_pos]
                new_state = ''.join(new_state)
                
                if new_state not in visited:
                    visited.add(new_state)
                    queue.append((new_state, neighbor, steps + 1))
        
        return -1
```
