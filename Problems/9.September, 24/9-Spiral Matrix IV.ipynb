You are given two integers `m` and `n`, which represent the dimensions of a matrix.

You are also given the `head` of a linked list of integers.

Generate an `m x n` matrix that contains the integers in the linked list presented in **spiral** order **(clockwise)**, starting from the **top-left** of the matrix. If there are remaining empty spaces, fill them with `-1`.

Return *the generated matrix*.

<br>

**Example 1:**

![ex1new](../../images/ex1new.jpg)

>**Input:** m = 3, n = 5, head = [3,0,2,6,8,1,7,9,4,2,5,5,0]<br>
>**Output:** [[3,0,2,6,8],[5,0,-1,-1,1],[5,2,4,9,7]]<br>
>**Explanation:** The diagram above shows how the values are printed in the matrix.<br>
>Note that the remaining spaces in the matrix are filled with -1.

**Example 2:**

![ex2new](../../images/ex2new.jpg)

>**Input:** m = 1, n = 4, head = [0,1,2]<br>
>**Output:** [[0,1,2,-1]]<br>
>**Explanation:** The diagram above shows how the values are printed from left to right in the matrix.<br>
>The last space in the matrix is set to -1.

<br>

**Constraints:**
- >1 <= m, n <= 10<sup>5</sup>
- >1 <= m * n <= 10<sup>5</sup>
- >The number of nodes in the list is in the range [1, m * n].
- >0 <= Node.val <= 1000

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def spiralMatrix(self, m: int, n: int, head) -> list[list[int]]:
        matrix = [[-1] * n for _ in range(m)]
    
        directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        dir_idx = 0
        row, col = 0, 0
        
        top, bottom, left, right = 0, m - 1, 0, n - 1
        
        current = head
        while current:
            matrix[row][col] = current.val
            current = current.next
            
            new_row = row + directions[dir_idx][0]
            new_col = col + directions[dir_idx][1]
            
            if not (top <= new_row <= bottom and left <= new_col <= right):
                if dir_idx == 0:
                    top += 1
                elif dir_idx == 1:
                    right -= 1
                elif dir_idx == 2:
                    bottom -= 1
                elif dir_idx == 3:
                    left += 1
                dir_idx = (dir_idx + 1) % 4
                
            row += directions[dir_idx][0]
            col += directions[dir_idx][1]
        
        return matrix
```
