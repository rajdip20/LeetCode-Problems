Given an `m x n` `board` of characters and a list of strings `words`, return *all words on the board*.

Each word must be constructed from letters of sequentially adjacent cells, where **adjacent cells** are horizontally or vertically neighboring. The same letter cell may not be used more than once in a word.

<br>

Example 1:

![search1](../images/search1.jpg)

>**Input:** board = [["o","a","a","n"],["e","t","a","e"],["i","h","k","r"],["i","f","l","v"]], words = ["oath","pea","eat","rain"]<br>
>**Output:** ["eat","oath"]

Example 2:

![search2](../images/search2.jpg)

>**Input:** board = [["a","b"],["c","d"]], words = ["abcb"]<br>
>**Output:** []

<br>

**Constraints:**
- >m == board.length
- >n == board[i].length
- >1 <= m, n <= 12
- >board[i][j] is a lowercase English letter.
- >1 <= words.length <= 3 * 10<sup>4</sup>
- >1 <= words[i].length <= 10
- >words[i] consists of lowercase English letters.
- >All the strings of words are unique.

```python
class Solution:
    def findWords(self, board: list[list[str]], words: list[str]) -> list[str]:
        def dfs(board, node, i, j, path, res):
            if node.isWord:
                res.add(path)
                node.isWord = False
            
            if i < 0 or i >= len(board) or j < 0 or j >= len(board[0]):
                return
            tmp = board[i][j]
            node = node.children.get(tmp)
            
            if not node:
                return
            board[i][j] = "#"
            
            for x, y in [(i-1, j), (i+1, j), (i, j-1), (i, j+1)]:
                dfs(board, node, x, y, path+tmp, res)
            board[i][j] = tmp

        class TrieNode:
            def __init__(self):
                self.children = {}
                self.isWord = False

        def addWord(root, word):
            node = root
            
            for letter in word:
                node = node.children.setdefault(letter, TrieNode())
            node.isWord = True

        root = TrieNode()
        for word in words:
            addWord(root, word)

        res = set()
        
        for i in range(len(board)):
            for j in range(len(board[0])):
                dfs(board, root, i, j, "", res)

        return list(res)
```
